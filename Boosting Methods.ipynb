```python
X = df.drop("class", axis=1)
```

```python
X.isnull().sum()
```

```python
X
```

```python
X = pd.get_dummies(X,drop_first=True)
```

```python
X
```

```python
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.15, random_state=101)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model = AdaBoostClassifier(n_estimators=1)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
print(classification_report(y_test, predictions))
```

```python
model.feature_importances_.argmax()
```

```python
X.columns[22]
```

```python
sns.countplot(x="odor", data=df, hue='class')
```

```python
error_rates = []

for n in range(1,20):
    model = AdaBoostClassifier(n_estimators=n)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    
    err = 1 - accuracy_score(y_test, preds)
    error_rates.append(err)
```

```python
plt.plot(range(1,20),error_rates)
```

```python
feats = pd.DataFrame(index=X.columns, data=model.feature_importances_,columns=['importance'])
```

```python
feats
```

```python
imp_feats = feats[feats['importance']>0]
```

```python
sns.barplot(x=imp_feats.index, data=imp_feats.sort_values("importance"), y="importance")
plt.xticks(rotation=90);
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_estimators':[50,100],
              'learning_rate' :[0.1,0.05,0.2],
              'max_depth':[3,4,5]} 
```

```python
gb_model = GradientBoostingClassifier()
```

```python
grid_search = GridSearchCV(gb_model,param_grid)
```

```python
grid_search.fit(X_train,y_train)
```

```python
preds = grid_search.predict(X_test)
```

```python
predictions
```

```python
grid_search.best_params_
```

```python
print(classification_report(y_test, preds))
```

```python
feat_import = grid_search.best_estimator_.feature_importances_
```

```python
imp_feat = pd.DataFrame(index=X.columns, data=feat_import,columns=['importance'])
```
